```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
data_file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(data_file)
```

## Player Count

* Display the total number of players


```python
total_players = df["SN"].nunique()
print("Total Players: " + str(total_players))
```

```python
df.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Number of unique items
unique_items = df["Item ID"].nunique()

# Average price()
average_price = round(df["Price"].mean(),2)

# Number of Purchases
number_of_purchases = df["Purchase ID"].count()

# Total Revenue
total_revenue = df["Price"].sum()

# Create summary dataframe- refer to goodreads activity day 1 exercise 11

pd.DataFrame({"Unique Items":[unique_items],"Average Price":average_price,"Number Of Purchases":number_of_purchases,"Total Revenue":total_revenue})
```

```python
df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Create a groupby dataframe object - Gender
gender_group = df.groupby("Gender")

# Count of each gender
gender_count = gender_group["SN"].nunique()

# Percentage of each gender - count of each gender / total number of players
percentage = gender_count/total_players
count = gender_count


# Create summary dataframe
pd.DataFrame({"Percentage by Gender":round(percentage*100,2),"Count by Gender":count})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# purchase count

purchase_count = gender_group["Purchase ID"].nunique()

# avg purchase price
purchase_average = gender_group["Price"].mean()

# total purchase value
purchase_total = gender_group["Price"].sum()

# Avg total purchase per person
average_total = purchase_total/count

pd.DataFrame({"Purchase Count":purchase_count,"Average Purchase Price":purchase_average,"Average Purchase Total":average_total,"Total Purchase Value":purchase_total})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# ted talks
# set bins

age_bins =  [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 99]
age_labels = [" <10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# cut

df["age_group"] = pd.cut(df["Age"], age_bins, labels = age_labels)

# groupby age bin

age_group = df.groupby("age_group")

age_count = age_group["SN"].nunique()

percentage_age = age_count/total_players
# get columns
pd.DataFrame({"Percentage by Age":round(percentage_age*100,2),"Count by Age":age_count})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# group by screenname
spender_group = df.groupby("SN")


purchase_count = spender_group["Purchase ID"].nunique()

# avg purchase price
purchase_average = spender_group["Price"].mean()

# total purchase value
purchase_total = spender_group["Price"].sum()


pd.DataFrame({"Purchase Count":purchase_count,"Average Purchase Price":purchase_average,"Total Purchase Value":purchase_total}).sort_values("Total Purchase Value",ascending=False)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_group = df.groupby(["Item ID","Item Name"])
item_price = item_group["Price"].mean()
item_count = item_group["Price"].count()
total_purchase = item_price*item_count

purchase_sort = pd.DataFrame({"Item Price":item_price,"Item Count":item_count,"Total Purchase":total_purchase})
purchase_sort.sort_values("Item Count",ascending=False)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting
